```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from random import shuffle
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv("loan_old.csv")
```

```python
data
```

```python
# i) Check for missing values

missing_values = data.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
#  ii) Check the type of each feature
print("Data Types:")
data.dtypes
```

```python
data['Credit_History'] = data['Credit_History'].astype(str)  # --?
data['Dependents'] = data['Dependents'].astype(str)
data.dtypes
```

```python
#  iii) Check whether numerical features have the same scale
num_cols = data.select_dtypes(include=[np.number]).columns  # exclude maxloan
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[num_cols])
scaled_data = pd.DataFrame(scaled_data, columns=num_cols)
print("num of cols ",num_cols)
print("Scaled Data:")
scaled_data.describe()
```

```python
# iv) Visualize a pairplot between numerical columns
sns.pairplot(data[num_cols])
plt.show()
```

```python
# remove missing values
data.dropna(inplace=True)
```

```python
data
```

```python
data = data.drop(columns=['Loan_ID'])
data
```

```python
# ii) Separate features and targets
X = data.drop(columns=["Max_Loan_Amount", "Loan_Status"])
y_amount = data["Max_Loan_Amount"]
y_status = data["Loan_Status"]
```

```python
print("status :\n ",y_status)
print("amount :\n",y_amount)
```

```python
# iii) Shuffle and split into training and testing sets
from sklearn.utils import shuffle

# Set a random seed for reproducibility
np.random.seed(42)

# Shuffle the data
indices = np.arange(X.shape[0])
np.random.shuffle(indices)

# Use iloc to index DataFrame using integer location (indices)
X = X.iloc[indices]
y_amount = y_amount.iloc[indices]
y_status = y_status.iloc[indices]

# Split into training and testing sets
X_train, X_test, y_amount_train, y_amount_test, y_status_train, y_status_test = train_test_split(
    X, y_amount, y_status, test_size=0.2, random_state=42
)
X_train
```

```python
print("x train: ",len(X_train))
print("x test: ",len(X_test))
print(X_train)

# print(X_test)
```

```python
# iv) Encode categorical features
from sklearn.preprocessing import LabelEncoder

categorical_features = ["Gender", "Married", "Dependents", "Education", "Credit_History", "Property_Area"]
for feature in categorical_features:
    le = LabelEncoder()
    X_train[feature] = le.fit_transform(X_train[feature])
    X_test[feature] = le.transform(X_test[feature])

print("---------train-------")
print( X_train)
```

```python
X_train.dtypes
```

```python
# v) Encode categorical targets
le_status = LabelEncoder()
y_status_train = le_status.fit_transform(y_status_train)
y_status_test = le_status.transform(y_status_test)
print("y_status_train \n",y_status_train)
print("y_status_test \n",y_status_test)
```

```python
# vi) Standardize numerical features
# Z-score normalization (mean = 0 and standard deviation=1)
from sklearn.preprocessing import StandardScaler
num_cols_numericalFeatures = [col for col in X_train.select_dtypes(include=[np.number]).columns ] 

scaler = StandardScaler()

X_train[num_cols_numericalFeatures] = scaler.fit_transform(X_train[num_cols_numericalFeatures])

X_test[num_cols_numericalFeatures] = scaler.transform(X_test[num_cols_numericalFeatures])

print(X_train[num_cols_numericalFeatures])

```

```python
# Step d: Fit a linear regression model amount
from sklearn.linear_model import LinearRegression

linear_reg_model = LinearRegression()

linear_reg_model.fit(X_train, y_amount_train)
y_amount_pred = linear_reg_model.predict(X_test)

print("y_amount_pred \n",y_amount_pred)

```

```python
from sklearn.metrics import r2_score


y_amount_test_pred = linear_reg_model.predict(X_test)

r2_test = r2_score(y_amount_test, y_amount_test_pred)

print("Testing R2 Score:", r2_test)
```

```python
# # Step f: Fit a logistic regression model 
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cost_function(y, y_pred):
    m = len(y)
    cost = -(1/m) * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
    return cost

def gradient_descent(X, y, y_pred, learning_rate):
    m = len(y)
    gradient = (1/m) * np.dot(X.T, (y_pred - y))
    return gradient * learning_rate

# Initialize weights and bias
theta = np.zeros(X_train.shape[1])
bias = 0
learning_rate = 0.01
iterations = 1000

for _ in range(iterations):
    z = np.dot(X_train, theta) + bias
    y_pred = sigmoid(z)
    cost = cost_function(y_status_train, y_pred)
    gradient = gradient_descent(X_train, y_status_train, y_pred, learning_rate)
    theta -= gradient
    bias -= (learning_rate / len(y_status_train)) * np.sum(y_pred - y_status_train)

y_status_pred = (y_pred > 0.5).astype(int)
print(len(y_status_pred))
print("Predicted Loan Status:")
print(y_status_pred)    


```

```python
# Step g:  accuracy of the model
def accuracy(y_true, y_pred):
    correct = np.sum(y_true == y_pred) # ++ if true
    total = len(y_true)
    return correct / total

y_status_pred = sigmoid(np.dot(X_test, theta) + bias)
y_status_pred = (y_status_pred > 0.5).astype(int)

accuracy_score = accuracy(y_status_test, y_status_pred)
print(f"Accuracy (Logistic Regression): {accuracy_score}")
```
